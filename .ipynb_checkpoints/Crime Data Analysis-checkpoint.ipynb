```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Dataset using csv

```python
df = pd.read_csv("../crime_ds.csv")
```

## Histogram of crime occurrences over time

```python
df['DATE OCC'] = pd.to_datetime(df['DATE OCC']) #Convert 'DATE OCC' to datetime type
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='DATE OCC', bins=50, kde=True)
plt.title('Distribution of Crime Occurrences Over Time')
plt.xlabel('Date of Occurrence')
plt.ylabel('Number of Crimes')
plt.show()
```

The above histogram shows the number of crimes reported on the y axis 
and the year  and month on which the crime was reported on the x axis 
It shows how the crimal activity was at its peak on June of 2022.

## Time series plot of crime occurrences

```python
df.set_index('DATE OCC', inplace=True)
df.resample('M').size().plot(legend=False)
plt.title('Monthly Crime Occurrences')
plt.xlabel('Year of Occurrence')
plt.ylabel('Number of Crimes')
plt.show()
```

## Bar Chart of Crime Type

```python
top_crime_types = df['Crm Cd Desc'].value_counts().nlargest(40).index

plt.figure(figsize=(12,12))
sns.countplot(y='Crm Cd Desc', data=df, order=top_crime_types)
plt.title('Number of reported case')
plt.ylabel('Type of Crime')
plt.show()
```

As with the plotted bar chart we can see that Vehicle stolen is the most common crime. There are more than 200 different crimes 
but I just plotted top 40 crimes with most reporting. After vehicle stolen other crimes like Battery and theft of identity
takes second and third place.
